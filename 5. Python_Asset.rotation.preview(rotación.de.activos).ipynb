# Rotación de activos ( Ventas netas/ total activos)

### En este apartado se analiza la rotación de activos mediante gráficos de violín que integran Ventas netas y Total activo para los 4 estados booleanos de la empresa (incubada viva, no incubada viva, incubada muerta y no incubada muerta)

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
datos= pd.read_csv('/Users/Usuario/Documents/Python Docs/UVigo374sindispersion.csv')
```

```python
Incubadas = datos[datos['Incubada'] == True]
iv = Incubadas[Incubadas['Estadodummy'] == False]
im = Incubadas[Incubadas['Estadodummy'] == True]

Noincubadas = datos[datos['Incubada'] == False]
niv= Noincubadas[Noincubadas['Estadodummy'] == False]
nim= Noincubadas[Noincubadas['Estadodummy'] == True]
```

### Violin plot de Total activo

```python
Y1 = iv['totalactivo_mean']
Y2 = im['totalactivo_mean']
Y3 = niv['totalactivo_mean']
Y4 = nim['totalactivo_mean']
  
# Initialise the subplot function using number of rows and columns
plt.rcParams["figure.figsize"]=(30, 30)


figure, axis = plt.subplots(2, 2)
plt.setp(axis,yticks=[100000, 200000, 400000, 600000, 900000])

plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15)
  
# For Sine Function
axis[0, 0].violinplot(Y1,showmedians=True)
axis[0, 0].set_title("Incubada viva",size=30)

  
# For Cosine Function
axis[0, 1].violinplot(Y2,showmedians=True)
axis[0, 1].set_title("Incubada muerta",size=30)
  
# For Tangent Function
axis[1, 0].violinplot(Y3,showmedians=True)
axis[1, 0].set_title("No incubada viva",size=30)
  
# For Tanh Function
axis[1, 1].violinplot(Y4,showmedians=True)
axis[1, 1].set_title("No incubada muerta",size=30)
  
# Combine all the operations and display


plt.show()
```

### Violin plot de Ventas netas

```python
Y1 = iv['Ventasnetas_mean']
Y2 = im['Ventasnetas_mean']
Y3 = niv['Ventasnetas_mean']
Y4 = nim['Ventasnetas_mean']
  
# Initialise the subplot function using number of rows and columns
plt.rcParams["figure.figsize"]=(20, 30)



figure, axis = plt.subplots(2, 2)
plt.setp(axis,yticks=[100000, 200000, 400000, 600000, 900000])



plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15)
  
# For Sine Function
axis[0, 0].violinplot(Y1,showmedians=True)
axis[0, 0].set_title("Incubada viva",size=30)

  
# For Cosine Function
axis[0, 1].violinplot(Y2,showmedians=True)
axis[0, 1].set_title("Incubada muerta",size=30)
  
# For Tangent Function
axis[1, 0].violinplot(Y3,showmedians=True)
axis[1, 0].set_title("No incubada viva",size=30)
  
# For Tanh Function
axis[1, 1].violinplot(Y4,showmedians=True)
axis[1, 1].set_title("No incubada muerta",size=30)
  
# Combine all the operations and display

plt.show()
```

# Rotación de activos: ventas / activos totales

```python
Y1 = iv['Ventasnetas_mean']
Y2 = im['Ventasnetas_mean']
Y3 = niv['Ventasnetas_mean']
Y4 = nim['Ventasnetas_mean']



figure, axis = plt.subplots(2, 2)
plt.setp(axis,yticks=[100000, 200000, 400000, 600000, 900000])



plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15)

  
# For Sine Function
axis[0, 0].violinplot(iv['Ventasnetas_mean'],showmedians=True)
axis[0, 0].violinplot(iv['totalactivo_mean'],showmedians=True)
axis[0, 0].set_title("Incubada viva",size=30)


  
# For Cosine Function
axis[0, 1].violinplot(im['Ventasnetas_mean'],showmedians=True)
axis[0, 1].violinplot(im['totalactivo_mean'],showmedians=True)
axis[0, 1].set_title("Incubada muerta",size=30)
  
# For Tangent Function
axis[1, 0].violinplot(niv['Ventasnetas_mean'],showmedians=True)
axis[1, 0].violinplot(niv['totalactivo_mean'],showmedians=True)
axis[1, 0].set_title("No incubada viva",size=30)
  
# For Tanh Function
axis[1, 1].violinplot(nim['Ventasnetas_mean'],showmedians=True)
axis[1, 1].violinplot(nim['totalactivo_mean'],showmedians=True)
axis[1, 1].set_title("No incubada muerta",size=30)





plt.show()
```

### Las ventas (Azul) y el Total activo (kaki): Las empresas no incubadas tienen unas ventas superiores a su Activo total...  

### En el gráfico Incubadas vivas se puede ver que ventas está equilibrada con el activo en empresas por debajo de 400.000 , desde ahí , las ventas decrecen considerablemente : esto podría indicar que las incubadoras son efectivas en empresas por debajo de ese valor, para empresas mayores parecen desaprovechado su potencial. 

### Las incubadas muertas tienen unas ventas menores que su activo en casi su totalidad

### Las No incubadas vivas tienen un nivel de ventas levemente superior que las incubadas, y se ven efectivas en empresas con un total activo mayor de los 400.000, por lo que parece irles mejor fuera de la incubadora desde los 200.000 de total activo

### El último es extraño, las no incubadas muertas tuvieron el valor de ventas superior a todas. 


### Se pueden plantear muchas hipótesis, como que las no incubadas podrían contar con un buen asesoramiento financiero y pagan por ello lo cual da mejores resultados a partir de cierto valor de la empresa.  También que las no incubadas muertas quizás fueron vendidas o cerradas en un buen momento y podría ser una flaqueza de las incubadoras.
